### Step 1: Load file
You are given a CSV file named: scores.csv

### Step 2: Pre-Process Data

* Load the numeric data from the CSV file into a NumPy array.
* Ignore the header row.
* Ignore the first column (names).

### Step 3: Process Each Row

For each student (each row in the array):

* Convert the row into a NumPy array.

    Calculate:

        - Total score

        - Average score

Print the results in this format:

Student 1 - Total: __ , Average: __

### Step 4: Conditional Processing

For each student:

    * If the average is ≥ 85, print "Excellent"
    * If the average is between 70 and 84, print "Good"
    * Otherwise, print "Needs Improvement"

```python
#Step 1 & 2
import numpy as np

scores = np.loadtxt("scores.csv", delimiter=",", skiprows=1, usecols=range(1, 4))
print(scores)


#Step 3
for i in range(len(scores)):
    row = scores[i]
    total = np.sum(row)
    average = np.mean(row)
    print(f"Student {i + 1}: Total = {total}, Average = {average}")


#Step 4
if average >= 85:
    print("Excellent")
elif average >= 70 and average < 85:
    print("Good")
else:
    print("Needs Improvement")


```
